```python
def add_content_col(df: pd.DataFrame) -> pd.DataFrame:
    content = df['title'] + " " + df['primary_category'] + " " + df['abstract']
    return pd.DataFrame({'paper_id': df['paper_id'], 'content': content})
```

```python
content_df = add_content_col(df)
```
